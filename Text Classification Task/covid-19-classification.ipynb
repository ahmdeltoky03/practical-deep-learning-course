```python
def get_category(x):
  if x in ['Extremely Positive', 'Positive']:
    return 1
  else :
    return 0

data['BCategory'] = data['Sentiment'].apply(lambda x : get_category(x))
data['BCategory'].value_counts()
```

```python
data['Sentiment'].value_counts()
```

```python
labels = {
    'Extremely Positive': 0,
    'Positive': 1,
    'Neutral': 2,
    'Negative': 3,
    'Extremely Negative': 4
}

data['MCategory'] = data['Sentiment'].map(labels)
data['MCategory'].value_counts()
```

```python
data
```

```python
tweet = 'advice talk to your neighbours family to exchange phone numbers create contact list with phone numbers of neighbours schools employer chemist gp set up online shopping accounts if poss adequate supplies of regular meds but not over orde'
len(tweet), len(tweet.split())
```

```python
data['TweetLength'] = data['OriginalTweet'].apply(lambda tweet : len(tweet))
data['SentenceLength'] = data['OriginalTweet'].apply(lambda tweet : len(tweet.split()))
data['Numbers'] = data['OriginalTweet'].apply(lambda x : len([i for i in x if i.isdigit()]))
data['Special'] = data['OriginalTweet'].apply(lambda x : len([i for i in x if i in '!@#$%^&*()']))
data['Links'] = data['OriginalTweet'].apply(lambda x : x.count('https:'))
```

```python
data
```

```python
data.columns
```

```python
X = data[["TweetLength", "SentenceLength", "Numbers", "Special", "Links"]]
y = data['BCategory']
```

```python
X
```

```python
y
```

```python
#Import Libraries
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

#Splitting data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=44, shuffle =True)
```

```python
#Splitted Data
print('X_train shape is ' , X_train.shape)
print('X_test shape is ' , X_test.shape)
print('y_train shape is ' , y_train.shape)
print('y_test shape is ' , y_test.shape)
```

```python
RandomForestClassifierModel = RandomForestClassifier(criterion = 'gini',n_estimators=500,max_depth=7,random_state=33) #criterion can be also : entropy
RandomForestClassifierModel.fit(X_train, y_train)
y_pred = RandomForestClassifierModel.predict(X_test)
ClassificationReport = classification_report(y_test,y_pred)
print('Classification Report is : \n', ClassificationReport )
```
