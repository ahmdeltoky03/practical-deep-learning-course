## Simple Code To Explain Data Augmentation

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import load_img, img_to_array, array_to_img
import os
import matplotlib.pyplot as plt
```

```python
datagen = ImageDataGenerator(
    horizontal_flip = True,
    rotation_range = 40,
    width_shift_range = .2,
    zoom_range = .2,
    shear_range = .2,
    brightness_range = (.5, 1.5)
)
```

```python
img = load_img("/content/image.png")
img = img_to_array(img)
img = img.reshape( (1, ) + img.shape)
```

```python
if not os.path.exists("augmented"):
  os.makedirs("augmented")
```

```python
i = 0
for batch in datagen.flow(img, batch_size = 1, save_to_dir = "augmented", save_prefix="image", save_format="png"):
  i+=1
  if i > 10:
    break
```

```python
for img in os.listdir("/content/augmented"):
  img = os.path.join("augmented", img)
  img = plt.imread(img) # img = load_img(img)
  plt.imshow(img)
  plt.axis("off")
  plt.show()
```
